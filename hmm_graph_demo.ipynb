```python
import numpy as np
from scipy.stats import dirichlet
from hmm_graph import draw_hmm
```

```python
# generate some random data
m = 5
lambda_ = np.random.randint(10, 200, m)
alpha = np.geomspace(100, 1, m)
gamma = np.vstack([dirichlet(np.roll(alpha, i)).rvs() for i in range(m)])
nze = np.random.randint(0, m*m, np.random.randint(0, m*m))
rx, ry = np.unravel_index(nze, (m, m))
gamma[rx, ry] = 0
delta = np.linalg.solve((np.eye(m) - gamma + 1).T, np.ones(m))
```

```python
fig, ax, graph = draw_hmm(lambda_, gamma, delta, layout='circular')
```
